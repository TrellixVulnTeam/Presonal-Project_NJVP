```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
import glob

path = './*'
file_list = glob.glob(path)  ## 폴더 안에 있는 모든 파일 출력

parts = [file for file in file_list if file.endswith('.csv')]  ## 특정 패턴의 파일만 출력
parts
```

```python
import re

def apply_regular_expression(text):
    hangul = re.compile('[^ ㄱ-ㅣ 가-힣]')  # 한글 추출 규칙: 띄어 쓰기(1 개)를 포함한 한글
    result = hangul.sub('', text)  # 위에 설정한 "hangul"규칙을 "text"에 적용(.sub)시킴
    return result
```

```python
from konlpy.tag import Okt
from collections import Counter
```

```python
okt = Okt()  # 명사 형태소 추출 함수
```

```python
for part in parts:
    df = pd.read_csv(part)
    print(f'{part}의 데이터 길이: ', len(df))
    titles = list(df['title'].unique())
    print("==========================================")
    for title in titles:
        print(title)
#         print(title," : ",len(df[df['title'] == title]))
        df_ = df[df['title'] == title]
        df_ = df_[['rating', 'text']]
        # 말뭉치 생성
        corpus = "".join(df_['text'].tolist())
        # 정규 표현식 적용
        apply_regular_expression(corpus)
        # 전체 말뭉치(corpus)에서 명사 형태소 추출
        nouns = okt.nouns(apply_regular_expression(corpus))
#         print(nouns)
        # 빈도 탐색
        counter = Counter(nouns)
        # 빈도수 top10
        print(counter.most_common(10))
        
#         break
        
    print("..................................................")
#     break
    
```
